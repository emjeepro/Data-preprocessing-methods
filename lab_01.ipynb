### Step №3. Add two new lines to the frame: information about the student Oleg Alekseevich Vasiliev, who received a grade of "3" in one of the disciplines (there are no grades for the rest of the disciplines) and about Alexander Nikolayevich Lovtsov, who does not have an assessment in one of the disciplines (fill in the rest of the grades arbitrarily). The numbers of the grade books for these students should be considered unknown (no values). Print the resulting frame and make sure that the lines were actually added.

### Этап №3. Добавить во фрейм две новые строки: информацию о студенте Васильеве Олеге Алексеевиче, который получил оценку «3» по одной из дисциплин (по остальным дисциплинам оценки отсутствуют) и о Ловцове Александре Николаевиче, не имеющем оценки по одной из дисциплин (остальные оценки заполнить произвольно). Номера зачетных книжек у этих студентов считать неизвестными (значения отсутствуют). Вывести полученный фрейм и убедиться, что строки действительно добавились.

```python
# understand that "new_line" is a dictionary

new_line = {'ФИО':'Васильев Олег Алексеевич', '№груп':'A', 'РКЧИС':'3'}
df = df.append(new_line, ignore_index=True)
new_line = {'ФИО':'Ловцов Александр Николаевич', '№груп':'B', 'СМАД':'4', 'МПОД':'3', 'ИЯАЦ':'5'}
df = df.append(new_line, ignore_index=True)
df
```

---

### Step №4. Fill in all the missing grades for all students with the value "2". Display the modified frame and check the result.

### Этап №4. Заполнить у всех студентов все отсутствующие оценки значением «2». Вывести измененный фрейм и проконтролировать результат.

```python
# understand the meaning of inplace=True/False

df.РКЧИС.fillna(2, inplace=True)
df.СМАД.fillna(2, inplace=True)
df.МПОД.fillna(2, inplace=True)
df.ИЯАЦ.fillna(2, inplace=True)
df
```

---

### Step №5. Add a new column to the frame containing information about the average score of each student for the current session. Display the modified frame and check the result.

### Этап №5. Добавить во фрейм новый столбец, содержащий информацию о среднем балле каждого студента за текущую сессию. Вывести измененный фрейм и проконтролировать результат.

```python
df["РКЧИС"] = pd.to_numeric(df["РКЧИС"])
df["СМАД"] = pd.to_numeric(df["СМАД"])
df["МПОД"] = pd.to_numeric(df["МПОД"])
df["ИЯАЦ"] = pd.to_numeric(df["ИЯАЦ"])

selected = df[['РКЧИС', 'СМАД', 'МПОД', 'ИЯАЦ']]
df['AVG'] = selected.mean(axis='columns')
df
```

```python
# dropping should be done based on an equation evaluating the AVG and not just judging by eye

# удалить студентов при условии, что средний балл меньше 3
# remove rows of students provided the GPA is less than 3

listfail = df[ df['AVG'] < 3 ]
fail = df[ df['AVG'] < 3 ].index
print(fail)
```
